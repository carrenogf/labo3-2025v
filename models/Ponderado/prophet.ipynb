```python
df_series = pd.pivot_table(sellin,index=["periodo",],columns=["product_id"],values=["tn"],aggfunc="sum",fill_value=0)
df_series.reset_index(inplace=True)
df_series.columns = df_series.columns.map(lambda x: x[1] if x[0] == "tn" else x[0])
df_series
```

```python
df_total = sellin.groupby("periodo")["tn"].sum().reset_index()
df_total.columns = ["periodo", "total_tn"]
df_total["periodo"] = pd.to_datetime(df_total["periodo"], format="%Y%m") + pd.offsets.MonthEnd(0)
df_total = df_total.sort_values(by='periodo')
df_total.set_index("periodo", inplace=True)
df_total = df_total.fillna(0)
# dfs = dfs.sort_values(by='periodo')
# dfs.set_index('periodo', inplace=True)
# dfs = dfs.fillna(0)


# 
df_total.head(3)
```

```python
from prophet import Prophet
import pandas as pd
import numpy as np

def generate_prophet_predictions(df, forecast_horizon=12,fourier_order=5):
    """
    Genera predicciones para cada columna/serie en el dataframe usando Prophet
    
    Args:
        df (pd.DataFrame): DataFrame con series temporales (columnas) y fechas como índice
        forecast_horizon (int): Número de periodos a predecir
        
    Returns:
        tuple: (df_predictions, df_models_info)
            - df_predictions: DataFrame con las predicciones
            - df_models_info: DataFrame con información de los modelos ajustados
    """
    df_predictions = pd.DataFrame()
    models_info = []
    
    for i, col in enumerate(df.columns, start=1):
        print(f"Ronda {i}: ajustando modelo para la serie '{col}'")
        # Preparar datos para Prophet
        df_series = df[[col]].reset_index()
        df_series.columns = ['ds', 'y']
        
        # Crear y ajustar modelo
        model = Prophet()
        model.add_seasonality(name='monthly', period=30.5, fourier_order=fourier_order)
        try:
            model.fit(df_series)

            # Crear fechas futuras
            future = model.make_future_dataframe(periods=forecast_horizon, freq='M')
            forecast = model.predict(future)

            # Extraer solo las predicciones futuras
            pred = forecast[['ds', 'yhat']].iloc[-forecast_horizon:].copy()
            pred.columns = ['ds', col]

            # Combinar predicciones
            if df_predictions.empty:
                df_predictions = pred
            else:
                df_predictions = df_predictions.merge(pred, on='ds')

            # Guardar info del modelo
            models_info.append({
                'Serie': col,
                'Modelo': 'Prophet',
                'Capacidad': 'Ajuste automático de tendencia estacional y festivos',
                'Última fecha entrenamiento': df_series['ds'].max()
            })

        except Exception as e:
            print(f"Error ajustando Prophet para {col}: {str(e)}")
            pred = pd.DataFrame({'ds': pd.date_range(start=df_series['ds'].max(), periods=forecast_horizon+1, freq='M')[1:], col: np.nan})
            df_predictions = df_predictions.merge(pred, on='ds', how='outer') if not df_predictions.empty else pred
            models_info.append({
                'Serie': col,
                'Modelo': f"Error: {str(e)}",
                'Capacidad': np.nan,
                'Última fecha entrenamiento': df_series['ds'].max()
            })

    return df_predictions.set_index('ds'), pd.DataFrame(models_info)
```

```python
# Ejecutar la función
df_predictions, df_models_info = generate_prophet_predictions(df_total, forecast_horizon=2)
```
