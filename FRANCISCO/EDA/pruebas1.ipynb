```python
cat_series = pd.pivot_table(df,index=["periodo",],columns=["cat1"],values=["tn"],aggfunc="sum",fill_value=0)
cat_series.reset_index(inplace=True)
cat_series.columns = cat_series.columns.map(lambda x: x[1] if x[0] == "tn" else x[0])
cat_series["total"] = cat_series[1:-1].sum(axis=1)
cat_series["periodo_dt"] = pd.to_datetime(cat_series["periodo"], format="%Y%m").dt.date

cat_series.tail(10)
```

```python
df_series = pd.pivot_table(df,index=["periodo",],columns=["product_id"],values=["tn"],aggfunc="sum",fill_value=0)
df_series.reset_index(inplace=True)
df_series.columns = df_series.columns.map(lambda x: x[1] if x[0] == "tn" else x[0])
df_series
```

```python
# Asegurar ventas positivas
ventas = df[df['tn'] > 0].copy()

# Calcular cuántos meses únicos con ventas tuvo cada producto
historial = (
    ventas.groupby('product_id')['periodo']
    .nunique()
    .reset_index(name='meses_con_ventas')
)

# Definir el total de meses del histórico disponible
total_meses = df['periodo'].nunique()

# Filtrar productos que:
# - tienen más de 3 meses con ventas
# - tienen el total de meses esperados (por ejemplo 36)
productos_estables = historial[historial['meses_con_ventas'] >= 24]['product_id'].tolist()

# Filtrar el DataFrame original para esos productos
df_productos_estables = df[df['product_id'].isin(productos_estables)]

print(f"🔎 Se encontraron {len(productos_estables)} productos estables (vendidos todos los meses del histórico).")
```

```python
df_series_pct = df_series.copy()
df_series_pct.iloc[:, 1:-1] = df_series_pct.iloc[:, 1:-1].div(df_series_pct.iloc[:, 1:-1].sum(axis=1), axis=0) * 100
```

```python
df_series_pct_mean = df_series_pct.iloc[:,1:-1].mean()
df_series_pct_mean = df_series_pct_mean.reset_index()
df_series_pct_mean.columns = ["product_id", "mean"]
df_series_pct_mean = df_series_pct_mean.sort_values(by="mean", ascending=False)
df_series_pct_mean = df_series_pct_mean[df_series_pct_mean["mean"]>0.1]
df_series_pct_mean = df_series_pct_mean.merge(productos, how="left", on="product_id")
df_series_pct_mean.sort_values(by="mean", ascending=False)
df_series_pct_mean["mean_acum"] = df_series_pct_mean["mean"].cumsum().round(2)

df_series_pct_mean
```

35 productos son casi el 50% de las ventas y 135 el 80%

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(df_series_pct_mean["product_id"].astype(str), df_series_pct_mean["mean"].cumsum(), color="skyblue")
plt.xlabel("product_id")
plt.ylabel("Mean acumulado (%)")
plt.title("Gráfico de barras acumuladas de la columna mean")
plt.xticks([], [])  # Oculta los labels del eje x por legibilidad
plt.tight_layout()
plt.show()
```

```python
# Pivot the data to get mean (%) by cat1 (category) as index and brand as columns
df_series_pct_mean_stacked = df_series_pct_mean.pivot_table(
    index="cat1", columns="brand", values="mean", aggfunc="sum", fill_value=0
)

ax = df_series_pct_mean_stacked.plot(kind="bar", stacked=True, figsize=(14, 7))
plt.ylabel("Mean (%)")
plt.title("Distribución de mean por categoría (cat1) y marca (brand)")
plt.legend(title="Marca", bbox_to_anchor=(1.05, 1), loc="upper left")
for c, brand in enumerate(df_series_pct_mean_stacked.columns):
    for i, v in enumerate(df_series_pct_mean_stacked[brand]):
        if v > 0:
            ax.text(i, df_series_pct_mean_stacked.iloc[i, :c+1].sum() - v/2, brand, ha='center', va='center', fontsize=8)
plt.tight_layout()
plt.show()
```
